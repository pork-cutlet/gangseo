### 주소 컬럼에서 '구'만 뽑아서 새 컬럼 만들기

```python
# len(df_station.loc[2, '주소']) > 1  ## True
```

```python
## 구 컬럼이 NaN 인 행만 추출

df_station[df_station['구'].isnull()]
```

```python
## 주소에 '-' 대신 아래의 값 넣기
## 신세계 본점: 서울특별시 중구 명동 소공로 63
## 우림골목시장 주차장: 서울특별시 중랑구 망우로62길 48

## https://developer-ping9.tistory.com/111
## Index 기반 접근방법 
## 596행, '주소'칼럼 값을 '서울특별시 중구 명동 소공로 63'로 변경
df_station.at[df_station[df_station['구'].isnull() == True].index[0],'주소'] = '서울특별시 중구 명동 소공로 63'    

## 615행, '주소'컬럼 값을 '서울특별시 중랑구 망우로62길 48'로 변경
df_station.at[df_station[df_station['구'].isnull() == True].index[1],'주소'] = '서울특별시 중랑구 망우로62길 48' 
```

## 4-1. 전기차 충전소 설치현황_20220417.csv

```python
df_so = pd.read_csv('../8. 내용보완해서_포트폴리오_만들자/data/4-1. 전기차 충전소 설치현황_20220417.csv')
df_so
```

```python
## 사용하고 싶은 컬럼만 뽑기

df_so_2 = df_so[['설치장소', '구', '급속충전기(대)', '완속충전기(대)', '지원차종']]
df_so_2
```

```python
pd.to_numeric(df_so_2['급속충전기(대)'], errors= 'coerce')
```

```python
pd.to_numeric(df_so_2['완속충전기(대)'], errors= 'coerce')
```

```python
df_so_2.info()
```

### 구별 급속충전기(대), 완속충전기(대) 비교

```python
df_dea = df_so_2.groupby('구')[['급속충전기(대)', '완속충전기(대)']].sum()
df_dea
```

```python
gu = df_dea.index.tolist()
rapid = df_dea['급속충전기(대)'].values.tolist()
slow = df_dea['완속충전기(대)'].values.tolist()
```

```python
len(gu)
```

#### 막대그래프

```python
import numpy as np
```

```python
# import seaborn as sns
# sns.set_theme(style='whitegrid')

# 주피터 노트북 환경설정
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

# 관련 라이브러리 임포트 
import matplotlib.font_manager as fm

#  한글글꼴로 변경
# plt.rcParams['font.family'] = '한글글꼴명'
plt.rcParams['font.size'] = 11.0
# plt.rcParams['font.family'] = 'batang'
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
matplotlib.rcParams['axes.unicode_minus'] = False

# plt.style.use('seaborn-whitegrid')
# plt.figure(figsize=(29,9))
```

```python
fig, ax = plt.subplots(figsize= (18, 6))
bar_width = 0.25


# 구가 25개 이므로, 0부터 24 위치를 기준으로 삼음
index = np.arange(len(gu))

# 각 별점별로 2개 어플의 bar 를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b_rapid = plt.bar(index, rapid, bar_width, alpha= 0.4, color= 'red', label= '급속충전기(대)')
b_slow = plt.bar(index + bar_width, slow, bar_width, alpha= 0.4, color= 'blue', label= '완속충전기(대)')

# x축 위치를 정 가운데로 조정하고, x축의 텍스트를 별점 정보와 매칭
plt.xticks(np.arange(bar_width/2, 25+bar_width/2, 1), gu)

# x축, y축 이름 및 범례 설정
plt.xlabel('전기차 충전기', size= 13, labelpad= 10)
plt.ylabel('개수', size= 13, labelpad= 10)
plt.legend()
plt.show()
```

### 구별 평균 값과 비교 그래프

```python
# 급속충전기(대) 평균 개수
avg_rapid = sum(rapid) / len(rapid)
avg_rapid
```

```python
# 완속충전기(대) 평균 개수
avg_slow = sum(slow) / len(slow)
avg_slow
```

```python
fig, ax = plt.subplots(figsize= (18, 6))
bar_width = 0.25


# 구가 25개 이므로, 0부터 24 위치를 기준으로 삼음
index = np.arange(len(gu))

# 각 별점별로 2개 어플의 bar 를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b_rapid = plt.bar(index, rapid, bar_width, alpha= 0.4, color= 'red', label= '급속충전기(대)')
b_slow = plt.bar(index + bar_width, slow, bar_width, alpha= 0.4, color= 'blue', label= '완속충전기(대)')

# x축 위치를 정 가운데로 조정하고, x축의 텍스트를 별점 정보와 매칭
plt.xticks(np.arange(bar_width/2, 25+bar_width/2, 1), gu)

## 충전기 평균선 그리기
## https://financedata.github.io/posts/matplotlib_hline_and_vline.html
plt.axhline(y= avg_rapid, color = 'r', linewidth= 1) # 급속 충전기 평균 값
plt.axhline(y= avg_slow, color= 'b', linewidth = 1) # 완속 충전기 평균 값
 
# x축, y축 이름 및 범례 설정
plt.xlabel('전기차 충전기', size= 13, labelpad= 10)
plt.ylabel('개수', size= 13, labelpad= 10)
plt.legend()
plt.show()
```

### 급속충전기(대)와 평균 대수와의 차이

```python
rapid_diff = [i - avg_rapid for i in df_dea['급속충전기(대)']] 
print(rapid_diff)
```

```python
## 색깔 구분을 위한 함수 (차이값이 0보다 크거나 같으면 red, 0보다 작으면 blue)
## color 참고: https://htmlcolorcodes.com/color-picker/
def diff_color(a) :
    if a >= 0 : 
        return '#FF5733' ## 'r'
    else : 
        return '#33BEFF' ## 'b'
    
## 함수 결과를 담을 새 리스트
diff_list = []

## 차이 리스트(rapid_diff) 의 요소에 함수 적용
for i in rapid_diff :
    x = diff_color(i)
    diff_list.append(x)

print(diff_list)
```

```python
# 참고: https://chancoding.tistory.com/112
fig = plt.figure(figsize= (18, 6))
ax = fig.add_subplot(111)

# 급속충전기(대) 그래프
ax.bar(gu, df_dea['급속충전기(대)'], color= diff_list)
ax.get_yaxis().get_major_formatter().set_scientific(False)

## 그래프 title 과 축 이름 지정
ax.set_title('< 급속충전기 대수의 평균과의 차 >', fontsize= 18, pad= 10)
ax.set_xlabel('충전기 대수', fontsize= 13)

plt.grid()
plt.show();
```

### 완속충전기(대)와 평균 대수와의 차이

```python
slow_diff = [i - avg_slow for i in df_dea['완속충전기(대)']] 
print(slow_diff)
```

```python
## 색깔 구분을 위한 함수 (차이값이 0보다 크거나 같으면 red, 0보다 작으면 blue)
def diff_color_slow(a) :
    if a >= 0 : 
        return '#FF5733' ## 'r'
    else : 
        return '#33BEFF' ## 'b'
    
## 함수 결과를 담을 새 리스트
diff_list_slow = []

## 차이 리스트(rapid_diff) 의 요소에 함수 적용
for i in slow_diff :
    x = diff_color_slow(i)
    diff_list_slow.append(x)

print(diff_list_slow)
```

```python
fig = plt.figure(figsize= (18, 6))
ax = fig.add_subplot(111)

# 완속충전기(대) 그래프
ax.bar(gu, df_dea['완속충전기(대)'], color= diff_list_slow)
ax.get_yaxis().get_major_formatter().set_scientific(False)

## 그래프 title 과 축 이름 지정
ax.set_title('< 완속충전기 대수의 평균과의 차 >', fontsize= 18, pad= 10)
ax.set_xlabel('충전기 대수', fontsize= 13)

plt.grid()
plt.show();
```
